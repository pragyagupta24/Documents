```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.histplot(iris, x='sepal_width', bins=15, kde=True)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
sns.histplot(iris, x='sepal_width', bins=15, kde=True, ax=ax)
plt.savefig('sepal_width_hist.png', dpi=300)
```

```python
sns.color_palette('rainbow')
```

```python
sns.set_palette('rainbow')
fig, ax = plt.subplots(figsize= (20,5), ncols=4)
sns.histplot(iris, x='sepal_width', bins=15, kde=True, ax=ax[0], color='red')
sns.histplot(iris, x='petal_width', bins=15, kde=True, ax=ax[1])
sns.histplot(iris, x='sepal_length', bins=15, kde=True, ax=ax[2])
sns.histplot(iris, x='petal_length', bins=15, kde=True, ax=ax[3])
```

```python
sns.set_palette('rainbow')
fig, ax = plt.subplots(figsize= (15,8), ncols=2, nrows=2)
sns.histplot(iris, x='sepal_width', bins=15, kde=True, ax=ax[0,0], color='red')
sns.histplot(iris, x='petal_width', bins=15, kde=True, ax=ax[0,1])
sns.histplot(iris, x='sepal_length', bins=15, kde=True, ax=ax[1,0])
sns.histplot(iris, x='petal_length', bins=15, kde=True, ax=ax[1,1])
plt.show()
```

```python
sns.barplot(iris, x='species', y='sepal_length')
```

```python
sns.barplot(iris, errorbar=None)
```
