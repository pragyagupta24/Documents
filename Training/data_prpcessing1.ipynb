```python
def clean_price_column(price):
    if isinstance(price, str):         # check if price variable is string
        price = price.replace(',', '') # remove comma
        price = price.strip()          # remove extra whitespace
        if price.isnumeric():
            price = int(price)
            return price
        if len(price) == 0:
            return np.nan
    return price        
```

```python
clean_price_column('23,902')
```

```python
df['price'].apply(clean_price_column)
```

```python
df['cleaned_price'] = df['price'].apply(clean_price_column)
```

```python
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating, str):
        rating = rating.split()[0]
    return rating    
```
